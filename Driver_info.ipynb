```python
acc_absence = pd.read_csv("Accident_Absence.csv")
```

```python
acc_absence.head(n=5)
```

```python
m= acc_absence.groupby('Absence_Number').size().sort_values(ascending=True)
freq_plot = m.plot(kind='bar',title='Bar chart for frequency of Absence_Number')
freq_plot.set_xlabel('Absence_Number')
freq_plot.set_ylabel('Frequency')
```

```python
n= acc_absence.groupby('Worked_Day_Off_Number').size().sort_values(ascending=True)
freq_plot1 = n.plot(kind='bar',title='Bar chart for frequency of Worked_Day_Off_Number')
freq_plot1.set_xlabel('Worked_Day_Off_Number')
freq_plot1.set_ylabel('Frequency')
```
